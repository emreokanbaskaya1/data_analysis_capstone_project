```python
yüksek_akustiklik = 0.6

# Verileri yüksek ve düşük akustiklik için filtrele
yüksek_akustiklik_grubu = df[df['Acousticness'] >= yüksek_akustiklik]
düşük_akustiklik_grubu = df[df['Acousticness'] < yüksek_akustiklik]

# Her grup için ortalama akış sayısını hesapla
yüksek_akustiklik_akış_ortalaması = yüksek_akustiklik_grubu['Stream'].mean()
düşük_akustiklik_akış_ortalaması = düşük_akustiklik_grubu['Stream'].mean()

# Sonuçları yazdır
print('Yüksek akustikliğe sahip sanatçılar için ortalama akış sayısı:', yüksek_akustiklik_akış_ortalaması)
print('Düşük akustikliğe sahip sanatçılar için ortalama akış sayısı:', düşük_akustiklik_akış_ortalaması)

#Düşük akustikliğe sahip parçaların ortalama akış sayısı daha yüksek.
```

```python
yüksek_dans = 0.6

# Verileri yüksek ve düşük dans edilebilirlik için filtrele
yüksek_dans_grubu = df[df['Danceability'] >= yüksek_dans]
düşük_dans_grubu = df[df['Danceability'] < yüksek_dans]

# Her grup için ortalama akış sayısını hesapla
yüksek_dans_akış_ortalaması = yüksek_dans_grubu['Stream'].mean()
düşük_dans_akış_ortalaması = düşük_dans_grubu['Stream'].mean()

# Sonuçları yazdır
print('Daha yüksek Dans Edilebilirlik değerine sahip sanatçılar için ortalama akış sayısı:', yüksek_dans_akış_ortalaması)
print('Daha düşük Dans Edilebilirlik değerine sahip sanatçılar için ortalama akış sayısı:', düşük_dans_akış_ortalaması)

#Daha yüksek Dans Edilebilirlik değerine sahip sanatçılar için ortalama akış sayısı daha yüksek.
```
